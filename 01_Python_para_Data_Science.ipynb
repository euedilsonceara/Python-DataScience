```python
#1) Nomes de variáveis podem começar com letras (a - z, A - Z) ou com caractere underscore (_)
#2) Os nomes de variáveis são case sensitive, ou seja, letras maiúsculas e minúsculas são consideradas

nome = "Edilson"
print(f"{nome}")

idade = 20
print(f"{idade}")

altura = 1.7
print(f"{altura}")

homem = True
print(f"{homem}")
```

##Criando funções / Função input(entrada) / Casting(Mudança de tipo de variável)

```python
#Toda entrada de dados através do input é uma string
#Para modificar o tipo de uma varável, faz-se um casting - int() str() float()

def saudacao_sem_parametros():
  nome = input("Digite o seu nome: ")
  idade = int(input("Digite a sua idade: "))
  print(f"Seja bem vindo(a) {nome}!! Você possui {idade} anos bem vividos")
saudacao_sem_parametros()

print()

def saudacao_com_parametros(nome,idade):
  print(f"Seja bem vindo(a) {nome}!! Você possui {idade} anos bem vividos")
saudacao_com_parametros("Felipe", 17)
```

##Operadores

```python
a = 10
b = 3

print("OPERADORES ARITMÉTICOS")
print(f"Soma:            {a+b}")
print(f"Subtração:       {a-b}")
print(f"Multiplicação:   {a*b}")
print(f"Exponenciação:   {a**b}")
print(f"Divisão:         {a/b}")
print(f"Divisão Inteira: {a//b}")
print(f"Módulo(Resto):   {a%b}")

print()

print("OPERADORES RELACIONAIS")
print(f"a igual a b:          {a == b}")
print(f"a diferente de b:     {a != b}")
print(f"a menor do que b:     {a < b}")
print(f"a maior do que b:     {a > b}")
print(f"a menor ou igual a b: {a <= b}")
print(f"a maior ou igual a b: {a >= b}")
```

##Listas e Fatiamento

```python
#Listas podem armazenar diferentes tipos de dados(inclusive outras listas) e são definidas entre colchetes []
#Listas são mutáveis
#Listas possuem índices, que iniciam a partir de 0

#No fatiamento A[i:j] o elemento i é incluído, mas o j não

idades = [10,18,15,25,17,26]
print(idades[2])

print("Slicing/Fatiamento: ")
print(idades[1:3])
print(idades[2:])
print(idades[:2])
print(idades[:])
```

##Operações em listas

```python
lista1 = [10,20,30,40,50,60]
lista2 = [70,80,90,100]

#Verifica se um valor existe ou não na lista
print(10 in lista1)
print(10 not in lista1)

#União/Concatenação de listas
lista3 = lista1 + lista2
print(lista3)

#Pega elemento da lista pelo índice
print(lista1[2])

#Tamanho da lista / Quantidade de elementos
print(len(lista1))

#Pega o maior elemento da lista
print(max(lista1))

#Pega o menor elemento da lista
print(min(lista1))

#Retorna o primeiro indice que ocorre um elemento
print(lista1.index(30))

#Retorna a quantidade de vezes que um elemento aparece
print(lista1.count(30))
```

##Métodos em listas

```python
#Ordenando elementos de uma lista ( .sort() - Altera a lista )
nomes = ["Edilson", "Ana", "Carlos", "Davi", "Marcos", "Italo"]
nomes.sort()
print(nomes)

idades = [25,12,45,23,87,57]
idades.sort()
print(idades)


#Adicionando elemento ao fim da lista ( .append() )
nomes.append("Maria")
print(nomes)

#Adicionando elemento em índice específico da lista ( .insert(indice,elemento) )
nomes.insert(0,"Abel")
print(nomes)


#Removendo elemento da lista ( .pop(indice) - Se não passar o índice, remove o último elemento)
nomes.pop()
print(nomes)


#Copiando uma lista ( .copy() )
nomes2 = nomes.copy()
print(nomes2)
nomes2.append("Chico")
print(nomes2)
```

##Laços/Loops

```python
idades = [10,18,15,25,17,26]

for idade in idades:
  print(idade)

print()

for i in range(0,10):
  print(i)
```

## Laços aninhados e Set



```python
dados = [[1,2,3,4,5], [5,6,7,8,9,10], [10,11,12,13,14,15]]
auxiliar = []

for lista in dados:
  for item in lista:
    auxiliar.append(item)
print(auxiliar)

#Removendo elementos duplicados
list(set(auxiliar))
```

##List Comprehensions

```python
#São formas concisas de criar listas em Python

print(list(range(10)))
print([i**2 for i in range(10)])

print(list(set([item for lista in dados for item in lista])))
```
